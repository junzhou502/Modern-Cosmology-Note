#### Exercise 1.2

Compare the cosmo age between $$\Omega_\Lambda = 0$$ and $$\Omega_\Lambda > 0$$ for a Euclidian universe with only matter and cosmo constant


```python
acoth = lambda x: 0.5*np.log((x+1)/(x-1))
f = lambda x: x**(-0.5)/3*2*(acoth(x**(-0.5)))


x = np.linspace(0.01,1,100)
y = f(x)

index = np.where(y>1)[0][0]


plt.plot(x, y)
plt.hlines(2/3,0,1, 'k')
plt.xlabel(r'$\Omega_\Lambda$')
plt.ylabel(r'$\tau (H_0^{-1})$')
plt.legend([r'$\Omega_\Lambda > 0$', r'$\Omega_\Lambda = 0$'])
plt.scatter(x[index], y[index],c ='k')
plt.text(x[index], y[index], str(np.round(x[index],2)), fontsize=15, c='r')
plt.title(r'Euclidiean Universe ($\Omega_m, \Omega_\Lambda$)')
```

#### Excercise 1.5

vaccume energy

consider scale that pc>>mc^2, calculate the vaccume energy density of the electron

```python
f = lambda x: x**2*np.sqrt(1+x**2)

I1 = sp.integrate.quad(f, 0, 10)
print(I1)
I2 = sp.integrate.quad(f, 0, 2.4e22)
print(I2)

C = 0.511**4/4/np.pi*1e-12
print('Integral 1 = ', C*I1[0])
print('Integral 2 = ', C*I2[0])
```
